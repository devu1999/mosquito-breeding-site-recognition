```python
import cv2
import os,sys
import random
import matplotlib.pyplot as plt
```

```python
dir = './ayush/Sem/Sem5/DIP/dip_project/dataset/rgb_with_puddle/'
```

```python
cnt = 0
for filename in os.listdir(dir):
    cnt+=1
    angle  = random.randint(0,180)
    img =  cv2.imread(dir+'images98.jpeg')
    (h, w) = img.shape[:2]
    scale = w/h
    center = (w / 2, h / 2)
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated90 = cv2.warpAffine(img, M, (h, w))
    cv2.imwrite(filename+'_'+str(cnt)+ '.jpg',rotated90)

    
```
